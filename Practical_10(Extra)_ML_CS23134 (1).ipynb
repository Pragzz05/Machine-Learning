```python
path="/content/Spotify_Songs_Dataset.csv"
df=pd.read_csv(path)
df.head()
```

```python
X = df[['danceability', 'energy']]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
sample_size = 200
np.random.seed(42)
sample_idx = np.random.choice(len(X_scaled), size=sample_size, replace=False)
X_sample = X_scaled[sample_idx]
```

```python
plt.figure(figsize=(12, 5))
dendrogram(linkage(X_sample, method='ward'))
plt.title('Dendrogram (Sample of 200 Songs)')
plt.xlabel('Songs')
plt.ylabel('Euclidean Distance')
plt.show()
```
